```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
log_text = open('./train_log_n.txt', 'r').read().lower()
log_strings = [log_string for log_string in log_text.split('\n\n') if log_string != '']
log_string_lists = [log_string.split('\n') for log_string in log_strings]
log_dicts = [{'time': int(log_string_list[0][:-1].split(' : ')[-1]),
              'epoch': int(log_string_list[1].split(' : ')[0].split('epoch ')[-1]),
              'loss': float(log_string_list[1].split(' : ')[1]),
              'accuracy': float(log_string_list[2].replace('%', '').split(' : ')[-1]) / 100,
              'recall': float(log_string_list[3].replace('%', '').split(' : ')[-1]) / 100,
              'precision': float(log_string_list[4].replace('%', '').split(' : ')[-1]) / 100,
              'f1': float(log_string_list[5].split(' : ')[-1])}
             for log_string_list in log_string_lists]
times = [log_dict['time'] for log_dict in log_dicts]
epochs = [log_dict['epoch'] for log_dict in log_dicts]
losses = [log_dict['loss'] for log_dict in log_dicts]
accuracies = [log_dict['accuracy'] for log_dict in log_dicts]
recalls = [log_dict['recall'] for log_dict in log_dicts]
precisions = [log_dict['precision'] for log_dict in log_dicts]
f1s = [log_dict['f1'] for log_dict in log_dicts]

print("Number of Logs :", len(log_strings))

fig, axes = plt.subplots(nrows=6, ncols=1, figsize=(15, 90), facecolor='lightslategrey')
fig.patch.set_alpha(1)
for ax, data, name, interval, significant_num in zip(axes,
                                    (times, losses, accuracies, recalls, precisions, f1s),
                                    ('Time (seconds)', 'Loss', 'Accuracy', 'Recall', 'Precision', 'F1 Score'),
                                    ((0, None), (0, None), (0, 1), (0, 1), (0, 1), (0, 1)),
                                    (max(times), min(losses), max(accuracies), max(recalls), max(precisions), max(f1s))):
    ax.grid()
    ax.patch.set_color((.4, .4, .4))
    ax.locator_params(axis='x', nbins=22)
    ax.locator_params(axis='y', nbins=10)
    ax.plot(epochs, data, color='cyan', alpha=0.5, linewidth=3)
    ax.set_xlim((0, None))
    ax.set_ylim(interval)
    ax.set_yticks(sorted(ax.get_yticks().tolist() + [float(significant_num)]))

    ax.set_title(name, family='sans-serif', fontsize=20, weight='bold')
    ax.set_xlabel('epoch', fontsize=15)
    ax.set_ylabel(name.lower(), fontsize=15)
```
